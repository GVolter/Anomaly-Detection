```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [5, 10, 2]
cov = [[3, 2, 2], [2, 10, 1], [2, 1, 2]]
data = np.random.multivariate_normal(mean, cov, size = 500)
data
```

```python
def plot_3d_data(ax, data, title, colors=['blue', 'red']):
    ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=colors[0], alpha=0.6)
    ax.set_title(title)
    ax.legend()

fig = plt.figure(figsize=(15, 10))

ax1 = fig.add_subplot(221, projection='3d')
plot_3d_data(ax1, data, "3D Dataset")

plt.tight_layout()
plt.show()
```

```python
mean_centered_data = data - np.mean(data, axis = 0) 
eigenvalues, eigenvectors = np.linalg.eigh(cov)

sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

pca_data = mean_centered_data @ eigenvectors[:, :2]

pca_data
```
